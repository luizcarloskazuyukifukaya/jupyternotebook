## panda dataframe

```python
df = pd.DataFrame(np.random.rand(6,4),
                 index=['one', 'two', 'three', 'four', 'five', 'six'],
                 columns=pd.Index(['A', 'B', 'C', 'D'],
                                 name='Genus'))
```

```python
df.shape
```

```python
df.info()
```

```python
df
```

```python
df.plot.bar()
```

```python
df.plot.barh(stacked=True)
```

```python
df_t = df.T
df_t
```

```python
df_t.plot.bar()
```

```python
df_t.plot.barh()
```

```python
df_t.plot.barh(stacked=True, alpha=0.5)
```

```python
df_t.plot.bar(stacked=True)
```
